**How to fix these columns**

We are going to consider that data in these columns which is present inside Upper Fence and Lower Fence

image.png

```python
Q1 = data.bmi.quantile(0.25)
Q3 = data.bmi.quantile(0.75)
IQR =  Q3-Q1     # Inter Quantile Range
UF = Q3+1.5*IQR
LF = Q1-1.5*IQR
# Consider only the data in the column present inside Upper and lower fence
data = data[ ( data.bmi >= LF ) & (data.bmi <=UF)]


Q1 = data.past_consultations.quantile(0.25)
Q3 = data.past_consultations.quantile(0.75)
IQR =  Q3-Q1     # Inter Quantile Range
UF = Q3+1.5*IQR
LF = Q1-1.5*IQR
# Consider only the data in the column present inside Upper and lower fence
data = data[ ( data.past_consultations >= LF ) & (data.past_consultations <=UF)]


Q1 = data.NUmber_of_past_hospitalizations.quantile(0.25)
Q3 = data.NUmber_of_past_hospitalizations.quantile(0.75)
IQR =  Q3-Q1     # Inter Quantile Range
UF = Q3+1.5*IQR
LF = Q1-1.5*IQR
# Consider only the data in the column present inside Upper and lower fence
data = data[ ( data.NUmber_of_past_hospitalizations >= LF ) & (data.NUmber_of_past_hospitalizations <=UF)]

Q1 = data.Anual_Salary.quantile(0.25)
Q3 = data.Anual_Salary.quantile(0.75)
IQR =  Q3-Q1     # Inter Quantile Range
UF = Q3+1.5*IQR
LF = Q1-1.5*IQR
# Consider only the data in the column present inside Upper and lower fence
data = data[ ( data.Anual_Salary >= LF ) & (data.Anual_Salary <=UF)]
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
# Condition to select only object data type column so we can convert these values to numerical value
for i in col:
  if data[i].dtypes=="object":  # select only object data type column
    data[i]=le.fit_transform(data[i])  # Apply label encoding to those columns
```

# Linear Regression Machine Learning Model Building

```python
# step -1 ( Dividing the data into Independent variable and dependent varible)

x=data.iloc[:, :-1 ]    # Independent variables

y=data["charges"]       # Dependent variable

# We are going to use independent variable to predict the value of dependent variable
```

```python
x
```

```python
y
```

```python
# Data splitting into Training data and Testing data
```

image.png

```python
from sklearn.model_selection import train_test_split

# Divide the data into 80:20 ( 80% data for Training and 20% data for Testing)
x_train , x_test , y_train , y_test = train_test_split ( x , y , test_size =0.20 ,random_state = 0)

# random_state will follow pattern to select some random values
```

```python
'''
Train means Training
Test means Testing

x means Independent variable
y means Dependent variable

x_train = Independent variable of Training data
y_train = Dependent variable of Training data

x_test = Independent variable of Testing data  ( age , bmi , no of steps etc )
y_test = Dependent variable of Testing data  ( charges )

'''
```

```python
# Build our Linear Regression model to predict charges
from sklearn.linear_model import LinearRegression
```

```python
# Give a name to your model and define which model it is
my_model =  LinearRegression()
```

```python
# Training the model
my_model.fit(x_train , y_train)  # Giving both question and answer to student for exam preparation
```

```python
# Test the model

# A question paper is going to be there with Indepndent variable of testing
model_answer = my_model.predict(x_test)
```

```python
x_test  # Independent variable of testing data
```

```python
model_answer  # Predicted charges values by my model
```

```python
y_test # Actual charges values
```

```python
# Evaluate the performance of model
from sklearn.metrics import *

result = r2_score( y_test  , model_answer)*100
result
```

```python
# Draw a Best fit line to compare how close the actual and predictec charges values are
sns.regplot( x=model_answer   , y= y_test)
plt.xlabel ("Predicted Charges values")
plt.ylabel ("Actual charges values")
plt.title("Best fit line of Linear Regression Model")
plt.show()
```
